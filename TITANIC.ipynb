```python
data = pd.read_csv("train.csv")
```

```python
def clean(data):
    data = data.drop(["Ticket", "Name", "Cabin", "PassengerId"], axis=1)
    
    cols = ["Parch", "SibSp", "Age", "Fare"]
    for col in cols:
      data[col].fillna(data[col].median(), inplace=True)
    
    data.Embarked.fillna("M", inplace=True)
    return data

data = clean (data)
test = clean (test)
```
